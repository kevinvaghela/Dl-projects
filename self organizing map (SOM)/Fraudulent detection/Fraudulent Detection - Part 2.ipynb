# Part 1 - Data Preprpcessing

## Importing Libaries

```python
import numpy as np 
import pandas as pd 

import tensorflow as tf
from tensorflow import keras

from keras.models import Sequential
from keras.layers import Dense
```

```python
dataset = pd.read_csv('../input/bank-data/bank_customers.csv')
```

```python
dataset.head()
```

## Let's divide into dependent(A1 to A14) and independent variable(Class)

```python
customers = dataset.iloc[:, 1:-1]
is_fraud = dataset.iloc[:, -1]
```

```python
customers.head()
```

```python
is_fraud[0:5]
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
Customers = sc.fit_transform(customers)
```

```python
np.random.seed(8)
tf.random.set_seed(8)
```

# Part 2 - Building the ANN
### Initializing the ANN

```python
classifier = Sequential()
```

### Adding the input layer and the hidden layer

```python
classifier.add(Dense(units=2, kernel_initializer='uniform', activation='relu', input_shape=(15,)))
```

### Adding the output layer

```python
classifier.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
```

### Compiling the ANN

```python
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics='accuracy')
```

### Fitting the ANN

```python
classifier.fit(customers, is_fraud, batch_size=1, epochs=2)
```

# Part 3 - making prediction (probability of fraudlent)

```python
y_pred = classifier.predict(customers)
```

### concatenate with the customersid so we can identify the customer

```python
y_pred = np.concatenate((dataset.iloc[:,0:1].values, y_pred), axis=1)
```

### let's sort the risk probability

```python
y_pred = y_pred[y_pred[:,1].argsort()]
```

```python
result = pd.DataFrame(y_pred)
```

```python
result.columns = ['CustomerID', 'Fraudlent probability']
```
